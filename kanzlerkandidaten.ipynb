```python
import pandas as pd
import numpy as np

# Text Datei Daten laden
with open("aussagen_der_kanzlerkandidaten.txt", "r", encoding="utf-8") as file:
    inhalt = file.read()
print(inhalt)
```

```python
doc = nlp(inhalt)
posCount = doc.count_by(spacy.attrs.POS)
print(posCount)
pos_labels = []
for pos in posCount.keys():
    pos_label = doc.vocab[pos].text
    pos_labels.append(pos_label)

pos_counts = list(posCount.values())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6)) 
plt.bar(pos_labels, pos_counts, color='skyblue')
```

```python
# !pip install wordcloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter

# Extrahiere die Wörter (ohne Stoppwörter)
words = [token.text for token in doc if not token.is_stop and not token.is_punct and not token.is_digit ]

# Zähle die Häufigkeiten der Wörter
word_counts = Counter(words)

# Sicherstellen, dass die Wortfrequenzen im richtigen Format vorliegen (Dictionary)
word_counts_dict = dict(word_counts)

# Konvertiere die Wortfrequenzen in einen flachen Text-String
flat_text = " ".join(words)

# Erzeuge die WordCloud
wordcloud = WordCloud(width=2500, height=2000, background_color="white").generate(flat_text)

# Darstellung der WordCloud
plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")  # Achsen entfernen
plt.show()
```
